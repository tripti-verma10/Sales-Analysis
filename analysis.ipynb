#### Convert column to corect data type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered']) #make int

all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) #make float

all_data.head()
```

### add a city column

```python
def get_city(address):
    if isinstance(address, str):  # Check if input is a string
        parts = address.split(',')
        return parts[1].strip() if len(parts) > 1 else "Unknown"
    return "Unknown"

def get_state(address):
    if isinstance(address, str):
        parts = address.split(',')
        return parts[2].strip().split(' ')[0] if len(parts) > 2 else "Unknown"
    return "Unknown"
all_data['Purchase Address'] = all_data['Purchase Address'].astype(str)

# Apply functions to extract city and state
all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' (' + get_state(x) + ')')

print(all_data.columns)
```

#### 4 . most product are most often sold together ?

```python
all_data.head()
```

```python

df = all_data[all_data['Order ID'].duplicated(keep=False)].copy()  

# Step 1: Use groupby and join the products
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

# Step 2: Drop duplicates and select columns (fixing typo in drop_duplicates)
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(30)
```

```python
from collections import Counter
from itertools import combinations

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key, value in count.most_common(10):
    print(key, value)
```

```python
prices = all_data.groupby('Product')['Price Each'].mean()

fig, ax1 = plt.subplots(figsize=(10, 6))

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g', alpha=0.6, label='Quantity Ordered')  
ax2.plot(products, prices, 'b-', marker='o', label='Price Each')

ax1.set_xlabel('Product')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price Each', color='b')

ax1.tick_params(axis='x', rotation=90)
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.set_title('Product Analysis: Quantity Ordered vs Price')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()


```
